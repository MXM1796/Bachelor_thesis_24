```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 500
```

```python
def plott_settings_default(plot_title, xlabel, ylabel, name_for_saving=0, legend_title=0, subtitle=0):
    plt.title(plot_title)
    plt.xlabel(xlabel, fontsize="12")
    plt.ylabel(ylabel, fontsize="12")
    if subtitle != 0: 
        plt.suptitle(subtitle)
    if legend_title!=0: 
        plt.legend(title=legend_title)
    if name_for_saving !=0: 
        plt.savefig(f'../mythesis_HQO_Reicherd/figs/{name_for_saving}', bbox_inches='tight')
    plt.show()
```

```python
# plot analog signals from the dataset csv from oscilloscope (lecroy / 500Mhz time resolution)

# load the dataset
data_00 = pd.read_csv('Analog signals/C12020-03-11 HV C00000.csv', delimiter=",", skiprows=5, header=None)
data_01 = pd.read_csv('Analog signals/C12020-03-11 HV C00001.csv', delimiter=",", skiprows=5, header=None)
data_02 = pd.read_csv('Analog signals/C12020-03-11 HV C00002.csv', delimiter=",", skiprows=5, header=None)
data_03 = pd.read_csv('Analog signals/C12020-03-11 HV C00003.csv', delimiter=",", skiprows=5, header=None)
data_04 = pd.read_csv('Analog signals/C12020-03-11 HV C00004.csv', delimiter=",", skiprows=5, header=None)
data_05 = pd.read_csv('Analog signals/C12020-03-11 HV C00005.csv', delimiter=",", skiprows=5, header=None)
data_06 = pd.read_csv('Analog signals/C12020-03-11 HV C00006.csv', delimiter=",", skiprows=5, header=None)
data_07 = pd.read_csv('Analog signals/C12020-03-11 HV C00007.csv', delimiter=",", skiprows=5, header=None)
data_08 = pd.read_csv('Analog signals/C12020-03-11 HV C00008.csv', delimiter=",", skiprows=5, header=None)
data_09 = pd.read_csv('Analog signals/C12020-03-11 HV C00009.csv', delimiter=",", skiprows=5, header=None)
data_10 = pd.read_csv('Analog signals/C12020-03-11 HV C00010.csv', delimiter=",", skiprows=5, header=None)
data_11 = pd.read_csv('Analog signals/C12020-03-11 HV C00011.csv', delimiter=",", skiprows=5, header=None)
data_12 = pd.read_csv('Analog signals/C12020-03-11 HV C00012.csv', delimiter=",", skiprows=5, header=None)
data_13 = pd.read_csv('Analog signals/C12020-03-11 HV C00013.csv', delimiter=",", skiprows=5, header=None)
data_14 = pd.read_csv('Analog signals/C12020-03-11 HV C00014.csv', delimiter=",", skiprows=5, header=None)
```

```python
# plot the analog signals
fig, axs = plt.subplots(2, 4, figsize=(15, 10))
fig.suptitle('Analog signals from the oscilloscope')

axs[0,0].plot(data_01[0], data_01[1])
axs[0,0].set_title('Bias current 31.2uA / trigger voltage: 600mV')
axs[0,1].plot(data_02[0], data_02[1])
axs[0,1].set_title('Bias current 33uA / trigger voltage: 600mV')
axs[0,2].plot(data_03[0], data_03[1])
axs[0,2].set_title('Bias current 35uA / trigger voltage: 600mV - latching')
axs[0,3].plot(data_04[0], data_04[1])
axs[0,3].set_title('Bias current 35uA / trigger voltage: 600mV - single')
axs[1,0].plot(data_05[0], data_05[1])
axs[1,0].set_title('Bias current 25uA / trigger voltage: 600mV - single')
axs[1,1].plot(data_12[0], data_12[1])
axs[1,1].set_title('Bias current 31.2uA / trigger voltage: 600mV (OD: 9.57)')
axs[1,2].plot(data_13[0], data_13[1])
axs[1,2].set_title('Bias current 31.2uA / trigger voltage: 600mV (OD: 9.57)')
axs[1,3].plot(data_14[0], data_14[1])
axs[1,3].set_title('Bias current 31.2uA / trigger voltage: 600mV (OD: 9.57)')
plt.show()
```

```python
#plot of the single analog signal 
plt.figure(figsize=(10, 5))
plt.plot(data_01[0]*10**9, data_01[1], label='Voltage curve of detected photon')
plt.axhline(y=0.600, color='r', linestyle='--', label='Trigger voltage')
xlabel = 'Time [ns]'
ylabel = 'Voltage [V]'
saving_name = 'HQO_20240725_analog_signal_31.2uA_tv_600mV.png'
plott_settings_default("",xlabel=xlabel, ylabel=ylabel, legend_title="", name_for_saving=saving_name)
```
