```python
# Efficiency calculation
# Constants
hbar = 1.054571817e-34  # Reduced Planck constant
c= 2.998e8  # Speed of light in vacuum (m/s)

# Wavelength in meters
lambda_IR = 780e-9

photons_per_second = lambda beam_pow, OD_value : beam_pow/(hbar * 2*np.pi* c/lambda_IR) *10**(-OD_value)


# Input Beam Power
beam_pow = 1.652e-6
OD_value = 6.0548

channel = 1
```

```python
Bias_sweep_files = os.listdir("Bias Sweep Variable OD")
current_limits_arr = []
photonrate_max_arr = []
OD_arr = np.zeros(len(Bias_sweep_files))

print(Bias_sweep_files)

for i, bias_file in enumerate(Bias_sweep_files): 
    data_sweep = np.loadtxt(f'Bias Sweep Variable OD\{bias_file}', skiprows=16, usecols=[0,9])
    # Berechne die Ableitungen
    dy = np.gradient(data_sweep[:,1])
    dx = np.gradient(data_sweep[:,0])
    steigungen = dy / dx

    # Finde die maximale Steigung
    max_Abfall = np.min(steigungen)
    max_Abfall_index = np.argmin(steigungen)    
    current_limits_arr.append(data_sweep[:,0][max_Abfall_index])
    photonrate_max_arr.append(np.max(data_sweep[:,1])*10**(-6))

    OD_arr[i] = float(str(bias_file).split(".txt")[0][40:-9])

    plt.plot(data_sweep[:,0],data_sweep[:,1]*10**(-6), label=str(bias_file).split(".txt")[0][37:-9] + "Limit " + str(data_sweep[:,0][max_Abfall_index]))
    plt.legend()    

title_bias_sweep = "Bias current vs Countrate"
plt.xlabel("Bias current [uA]")
plt.ylabel("Counts [MHz]")
# plt.savefig(f"{title_bias_sweep}.png")
plt.show()

plt.scatter(current_limits_arr, photons_per_second(beam_pow, OD_arr)*10**(-6), marker="+")
plt.ylabel("Photonrate [MHz]")
plt.xlabel("Max. Bias current [uA]")
title_bias_limit_OD = "Bias current Limit vs OD"
# plt.savefig(f"Bias Sweep Variable OD\{title_bias_limit_OD}.png")
plt.show()


plt.scatter(photons_per_second(beam_pow, OD_arr)*10**(-6), current_limits_arr, marker="+")
plt.xlabel("Photonrate [MHz]")
plt.ylabel("Max. Bias current [uA]")
title_bias_limit_OD = "Bias current Limit vs OD"
# plt.savefig(f"Bias Sweep Variable OD\{title_bias_limit_OD}.png")
plt.show()

photonrate_max_arr = np.array(photonrate_max_arr)
plt.scatter(photons_per_second(beam_pow, OD_arr)*10**(-6), photonrate_max_arr*5, marker="+")
sde_arr = [(photonrate_max_arr*5)*10**(6)/photons_per_second(beam_pow, OD_arr)]
print(np.sort(sde_arr))
plt.xlabel("Photonrate Input [MHz]")
plt.ylabel("Photonrate Detected [MHz]")
title_bias_limit_OD = "Bias current Limit vs OD"
# plt.savefig(f"Bias Sweep Variable OD\{title_bias_limit_OD}.png")
plt.show()

plt.scatter((photonrate_max_arr*5)*10**(6), sde_arr[0])
plt.xlabel("Photonrate Input [MHz]")
plt.ylabel("SDE [%]")
title_bias_limit_OD = "Photon Inputrate dependend SDE"
# plt.savefig(f"Bias Sweep Variable OD\{title_bias_limit_OD}.png")
plt.show()
```
